```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style = "whitegrid")
```

```python
df = pd.read_csv('/content/datos_campistas - Sheet1.csv')
df
```

```python
df.info()
```

```python
df.describe(include="all")
df.isnull().sum()
```

```python
df['estatura (cm)'] = df['estatura (cm)'].fillna(df['estatura (cm)'].mean())
df.isnull().sum()
```

```python
df['color de cabello'].value_counts()
```

```python
df['color de cabello'] = df['color de cabello'].fillna(df['color de cabello'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df = df.drop('mes de nacimiento', axis = 1)
```

#Limpieza de nombres y formatos

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.columns
```

```python
df['¿trabaja?'].value_counts()
```

```python
df['¿trabaja?']=df['¿trabaja?'].str.lower().str.strip().replace({"sí":"si"})
df['¿trabaja?'].value_counts()
```

```python
df['color_de_ojos'].value_counts()
```

```python
df['color_de_ojos'] = df['color_de_ojos'].str.lower().str.strip().replace({"café":"cafe","verde":"verdes"})
df['color_de_ojos'].value_counts()
```

```python
df['ciudad_de_nacimiento'].value_counts()
```

```python
df['ciudad_de_nacimiento'] = df['ciudad_de_nacimiento'].str.lower().str.strip().replace({"medellín":"medellin"})
```

```python
df['ciudad_de_nacimiento'].value_counts()
```

```python
df['género_musical_favorito'].value_counts()
```

```python
df['género_musical_favorito'] = df['género_musical_favorito'].str.lower().str.strip().replace({"grossover":"cross over"})
df['género_musical_favorito'].value_counts()
```

```python
df['profesión'] = df['profesión'].str.lower().str.strip().replace({"ingeniera":"ingeniero","ingeniera de sistemas":"ingeniero de sistemas"})
df['profesión'].value_counts()
```

```python
columnas_categoricas = ['sexo','rh','color_de_cabello','color_de_ojos','¿trabaja?']
df_encoded = pd.get_dummies(df,columns=columnas_categoricas,drop_first=True)
df_encoded.head(10)
```

```python
df_encoded.info()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df_encoded['peso_(kg)'])
plt.title("Boxplot variables numericas")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df_encoded['talla_de_zapato_(cm)'])
plt.title("Boxplot variables numericas")
plt.show()
```

```python
def detectar_outliers(col):
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  outliers = df[(df_encoded[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
  return outliers

outliers_estatura = detectar_outliers("talla_de_zapato_(cm)")
print("Outliers estatura: ")
print(outliers_estatura)
```

```python
variables_numericas = list(df_encoded.columns)
print(variables_numericas)

tipo_datos = ['int64','float64']
#Para sacar las columnas que solo son numericas se puede utilizar
variables_numericas1 = df_encoded.select_dtypes(include=['int64','float64'])
columnas_numericas = list(variables_numericas1)

for i in variables_numericas:
  print(df[i].dtypes)

  if df[i].dtypes in tipo_datos:
    outliers_gen = detectar_outliers(i)
    print("Outliers estatura: ")
    print(outliers_gen)
```

```python
df[columnas_numericas].describe()
from sklearn.preprocessing import MinMaxScaler, StandardScaler
df_numerico = df[columnas_numericas].copy()
```

```python
# Min Max
scaler_minmax = MinMaxScaler()
df_minmax = pd.DataFrame(scaler_minmax.fit_transform(df_numerico),columns=columnas_numericas)
```

```python
# Z-score
scaler_z= StandardScaler()
df_zcore = pd.DataFrame(scaler_z.fit_transform(df_numerico),columns=columnas_numericas)
```

```python
# Graficas
fig, axs = plt.subplots(3,5,figsize=(18,12))
for i, col in enumerate(columnas_numericas):
  sns.histplot(df_numerico[col], kde = True, ax = axs[0,i], color = 'skyblue')
  axs[0,i].set_title(f"Original - {col}")

  sns.histplot(df_minmax[col], kde = True, ax = axs[1,i], color = 'green')
  axs[1,i].set_title(f"Min Max - {col}")

  sns.histplot(df_zcore[col], kde = True, ax = axs[2,i], color = 'violet')
  axs[2,i].set_title(f"Z- score - {col}")

plt.tight_layout()
plt.show()
```

```python
# Curtosis 19/06/2025
from scipy.stats import kurtosis

for col in columnas_numericas:
  valor_curtosis = kurtosis(df[col].dropna())
  print(f"Curtosis de {col} : {valor_curtosis: .3f}")

for col in columnas_numericas:
  print(f"\n Percentiles para {col} : ")
  for percentile in [10,25,50,75,90]:
    valor = df[col].quantile(percentile / 100)
    print(f" Percentil {percentile} : {valor}")
```

```python
from scipy.stats import pearsonr, spearmanr, kendalltau

#  Correlación de Pearson
coeficiente, p_valor = pearsonr(df['edad'], df['peso_(kg)'])
print("Correlación de Pearson:", coeficiente, "p-valor:", p_valor)
#fig, axs = plt.subplots(1,1,figsize=(9,6))
sns.histplot(df[['edad','peso_(kg)']], kde = True,  color = 'skyblue')
plt.tight_layout()
plt.show()
#axs[coeficiente,p_valor].set_title(f"Original - ")

plt.figure(figsize=(10, 7))
sns.barplot(
    data=df,
    x='edad',
    y='peso_(kg)',
    hue='edad',
    )
plt.title('Relación entre edad y peso')
plt.xlabel('Edad')
plt.ylabel('Peso')

plt.legend(title='Leyenda', loc='upper right')

# Ajustar distribución final de los gráficos
plt.tight_layout()
plt.show()


#  Correlación de Spearman
coeficiente, p_valor = spearmanr(df['edad'], df['peso_(kg)'])
print("Correlación de Spearman:", coeficiente, "p-valor:", p_valor)

#  Correlación de Kendall
coeficiente, p_valor = kendalltau(df['edad'], df['peso_(kg)'])
print("Correlación de Kendall:", coeficiente, "p-valor:", p_valor)
```
